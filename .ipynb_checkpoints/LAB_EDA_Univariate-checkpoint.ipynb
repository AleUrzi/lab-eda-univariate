Part 1: Understanding Product Categories

**Business Question**: What are the most popular product categories on Amazon UK, and how do they compare in terms of listing frequency?

1. **Frequency Tables**:
    - Generate a frequency table for the product `category`.
    - Which are the top 5 most listed product categories?

2. **Visualizations**:
    - Display the distribution of products across different categories using a bar chart. *If you face problems understanding the chart, do it for a subset of top categories.*
    - For a subset of top categories, visualize their proportions using a pie chart. Does any category dominate the listings?


```python
import numpy as np
import pandas as pd

amazon_df = pd.read_csv("amz_uk_price_prediction_dataset.csv")

amazon_df.head()
```

```python
#Generate a frequency table for the product `category`.
#Which are the top 5 most listed product categories?

freq_table_cat = amazon_df["category"].value_counts().sort_values(ascending=False)
most_freq_table = freq_table_cat.head(5)
most_freq_table
```

```python
#2. **Visualizations**:
#Display the distribution of products across different categories using a bar chart. *If you face problems understanding the chart, do it for a subset of top categories.*
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

fig, axes = plt.subplots()
sns.barplot(x=most_freq_table.index, y=most_freq_table.values, palette="Set3", hue=most_freq_table.index, legend=False,
            order=['Sports & Outdoors', 'Beauty', 'Handmade Clothing, Shoes & Accessories', 'Bath & Body', 'Birthday Gifts'], ax=axes)
axes.set_ylabel("Frequency")
axes.set_xlabel("Category")
plt.xticks(rotation=45, ha='right')

plt.show()
```

```python
#For a subset of top categories, visualize their proportions using a pie chart. Does any category dominate the listings? answer: Sports and Outdoors

most_freq_table.plot.pie(autopct='%1.1f%%', startangle=50, colors=sns.color_palette("Set3"))
```

### Part 2: Delving into Product Pricing

**Business Question**: How are products priced on Amazon UK, and are there specific price points or ranges that are more common?

1. **Measures of Centrality**:
    - Calculate the mean, median, and mode for the `price` of products.
    - What's the average price point of products listed? How does this compare with the most common price point (mode)?

2. **Measures of Dispersion**:
    - Determine the variance, standard deviation, range, and interquartile range for product `price`.
    - How varied are the product prices? Are there any indicators of a significant spread in prices?

3. **Visualizations**:
    - Is there a specific price range where most products fall? Plot a histogram to visualize the distribution of product prices. *If its hard to read these diagrams, think why this is, and explain how it could be solved.*.
    - Are there products that are priced significantly higher than the rest? Use a box plot to showcase the spread and potential outliers in product pricing. 

---

```python
#1. **Measures of Centrality**:
#Calculate the mean, median, and mode for the `price` of products.
#What's the average price point of products listed? How does this compare with the most common price point (mode)?


mean = amazon_df["price"].mean()
median = amazon_df["price"].median()
mode = amazon_df["price"].mode()[0]

print(f"the mean is: {mean}, the median is: {median}, the mode is: {mode}")

#THE AVERAGE PRICE IS MUCH HIGHER THAN THE MODE. This might indicate the presence of outliers with very high prices 
```

```python
#2. **Measures of Dispersion**:
#Determine the variance, standard deviation, range, and interquartile range for product `price`.
#How varied are the product prices? Are there any indicators of a significant spread in prices?

# Determine the variance, standard deviation, range, and interquartile range for product price
variance_price = amazon_df["price"].var()
std_dev_price = amazon_df["price"].std()
min_price = amazon_df["price"].min()
max_price = amazon_df["price"].max()
range_price = max_price - min_price
quantiles_price = amazon_df["price"].quantile([0.25, 0.5, 0.75])
iqr_price = quantiles_price[0.75] - quantiles_price[0.25]

print(f"Variance: {variance_price}")
print(f"Standard Deviation: {std_dev_price}")
print(f"Range: {range_price}")
print(f"Interquartile Range (IQR): {iqr_price}")
print(f"Quantiles: {quantiles_price}")
```

```python
amazon_df["price"].value_counts()
```

```python
#3. **Visualizations**:
#Is there a specific price range where most products fall? Plot a histogram to visualize the distribution of product prices.
sns.histplot(amazon_df["price"], kde=True, bins=20, color="salmon")
```

```python
#If its hard to read these diagrams, think why this is, and explain how it could be solved.*.
# Convert prices into logaritmic scale

amazon_df["log_price"] = np.log1p(amazon_df["price"])  # log1p is used to avoid log(0)

sns.histplot(amazon_df["log_price"], kde=True, bins=20, color="salmon")
plt.xlabel("Log of Price")
plt.ylabel("Frequency")
plt.title("Histogram of Log-Transformed Product Prices")
plt.show()
```

```python
#Are there products that are priced significantly higher than the rest? 
#Use a box plot to showcase the spread and potential outliers in product pricing. 
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.boxplot(x=amazon_df["price"], color="skyblue")
plt.xlabel("Price")
plt.title("Box Plot of Product Prices")
plt.show()


#Many outliers, high variability of prices
```

### Part 3: Unpacking Product Ratings

**Business Question**: How do customers rate products on Amazon UK, and are there any patterns or tendencies in the ratings?

1. **Measures of Centrality**:
    - Calculate the mean, median, and mode for the `rating` of products.
    - How do customers generally rate products? Is there a common trend?

2. **Measures of Dispersion**:
    - Determine the variance, standard deviation, and interquartile range for product `rating`.
    - Are the ratings consistent, or is there a wide variation in customer feedback?

3. **Shape of the Distribution**:
    - Calculate the skewness and kurtosis for the `rating` column. 
    - Are the ratings normally distributed, or do they lean towards higher or lower values?

4. **Visualizations**:
    - Plot a histogram to visualize the distribution of product ratings. Is there a specific rating that is more common?


```python
#Calculate the mean, median, and mode for the rating of products.
#How do customers generally rate products? Is there a common trend?

mean = amazon_df["stars"].mean()
median = amazon_df["stars"].median()
mode = amazon_df["stars"].mode()[0]

print(f"the mean is: {mean}, the median is: {median}, the mode is: {mode}")

#Most prodcucts are rated 0
```

```python
#Determine the variance, standard deviation, and interquartile range for product rating.
#Are the ratings consistent, or is there a wide variation in customer feedback?
variance_stars = amazon_df["stars"].var()
std_dev_stars = amazon_df["stars"].std()
min_stars = amazon_df["stars"].min()
max_stars = amazon_df["stars"].max()
range_stars = max_price - min_price
quantiles_stars = amazon_df["stars"].quantile([0.25, 0.5, 0.75])
iqr_stars = quantiles_stars[0.75] - quantiles_stars[0.25]

print(f"Variance: {variance_stars}")
print(f"Standard Deviation: {std_dev_stars}")
print(f"Range: {range_stars}")
print(f"Interquartile Range (IQR): {iqr_stars}")
print(f"Quantiles: {quantiles_stars}")
```

```python
#Shape of the Distribution: Calculate the skewness and kurtosis for the rating column.
#Are the ratings normally distributed, or do they lean towards higher or lower values?

skewness_price = amazon_df["stars"].skew()
kurtosis_price = amazon_df["stars"].kurtosis()

skewness_price, kurtosis_price

#positive skeweness but very close to 0, indicating that the distribution of ratings is nearly symmetrica
#negative kurtosis: normal distribution
```

```python
#Plot a histogram to visualize the distribution of product ratings. Is there a specific rating that is more common?

sns.histplot(amazon_df["stars"], kde=True, bins=20, color="salmon")

#rating 0 is the most common
```
